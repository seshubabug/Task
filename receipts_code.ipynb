```python
import sys
import cv2 
import os
import numpy as np
import pytesseract
import matplotlib.pyplot as plt

import datetime
from datetime import date
import re
import datefinder
%matplotlib inline
```

```python
images = []

path = 'E:\\CEN\\jobs\\job task'

folders = os.listdir(path)
for folder in folders:
    #print(folder)
    folderpath = os.path.join(path,folder)
    if folderpath.endswith('Receipts'):
        print(folderpath)
        files = os.listdir(folderpath)
        for each in files:
            datapath = os.path.join(folderpath,each)
            print(datapath)
            img1 = cv2.imread(datapath)
            images.append(img1)
    
    
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\\Program Files\\Tesseract-OCR\\tesseract'

plt.figure(figsize=(20, 15))
plt.imshow(images[2])
print(images[2].shape)
print(len(images))
text1 = pytesseract.image_to_string(images[2])
print(text1)
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\\Program Files\\Tesseract-OCR\\tesseract'
count = 0
for i in range(len(images)):
    #print(i)
   
    
    data1=re.search(r'(Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)\s\d{2},\s\d{4}', pytesseract.image_to_string(images[i]))
    data2=re.search(r'\d{1,2}-(Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember))-\d{4}', pytesseract.image_to_string(images[i]))
    data3=re.search(r'[\d]{1,2}/[\d]{1,2}/[\d]{2,4}', pytesseract.image_to_string(images[i]))
    data4=re.search(r'[\d]{1,2}\.[\d]{1,2}\.[\d]{2,4}', pytesseract.image_to_string(images[i]))
    data5=re.search(r'\d{1,2}/(Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember))/\d{4}', pytesseract.image_to_string(images[i]))
    data6=re.search(r'[\d]{1,2}-[\d]{1,2}-[\d]{4}', pytesseract.image_to_string(images[i]))
    
    data7=re.search(r'(Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)\d{2}\'\d{2,4}', pytesseract.image_to_string(images[i]))
    data8=re.search(r'[\d]{4}-[\d]{1,2}-[\d]{1,2}', pytesseract.image_to_string(images[i]))
    
    if data1:
        #print("1-",data1.group(0))
        #print("Date:","MMM DD, YYYY")
        matches = datefinder.find_dates(data1.group(0))
        for match in matches:
            print(match)
            count = count+1
            
    elif data2:
        #print("2-",data2.group(0))
        #print("Date:","DD-MMM-YYYY")
        matches = datefinder.find_dates(data2.group(0))
        for match in matches:
            print(match)
            count = count+1
            
    elif data3:
        #print("3-",data3.group(0))
        #print("Date:","DD/MM/YYYY")
        matches = datefinder.find_dates(data3.group(0))
        for match in matches:
            print(match)
            count = count+1
            
    elif data4:
        #print("4-",data4.group(0))
        #print("Date:","DD.MM.YYYY")
        matches = datefinder.find_dates(data4.group(0))
        for match in matches:
            print(match)
            count = count+1
            
    elif data5:
        #print("5-",data5.group(0))
        #print("Date:","DD/MMM/YYYY")
        matches = datefinder.find_dates(data5.group(0))
        for match in matches:
            print(match)
            count = count+1
            
    elif data6:
        #print("6-",data6.group(0))
        #print("Date:","DD-MM-YYYY")
        matches = datefinder.find_dates(data6.group(0))
        for match in matches:
            print(match)
            count = count+1
            
    elif data7:
        #print("7-",data7.group(0))
        #print("Date:","MMM DD'YYYY")
        matches = datefinder.find_dates(data7.group(0))
        for match in matches:
            print(match)
            count = count+1
            
    elif data8:
        #print("8-",data8.group(0))
        #print("Date:","YYYY-MM-DD")
        matches = datefinder.find_dates(data8.group(0))
        for match in matches:
            print(match)
            count = count+1
    else:
        print('null')

       
```

```python
print("correct matches =", count)

Accuracy = (count/(len(images)))* 100
print("Accuracy =", Accuracy)
```
